## las Variables
1. CP --> Costo de producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChart
<img src="https://raw.githubusercontent.com/alexandroaguilar128/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20cambio%20(2).jpeg" width="350"  height="500">

```python
# Para cuando es verdadero (si)
# Calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CD = float(input("Ingrese el dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Alexandro Aguilar NL = 1")
```

```python
# Para cuando es Falso (no)
# Calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CD = float(input("Ingrese el dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Alexandro Aguilar NL = 1")
```
